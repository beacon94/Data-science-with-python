```python
from sklearn.metrics import ConfusionMatrixDisplay
import numpy as np
import matplotlib.pyplot as plt

# Format: (TP, FP, FN, TN)
conf_matrices = {
    'Decision Tree': {
        'SMOTE': (3405, 10214, 8755, 60100),
        'Under Sampling': (7170, 26973, 5192, 43139)
    },
    'Random Forest': {
        'SMOTE': (1682, 2864, 10332, 67596),
        'Under Sampling': (8563, 24371, 3670, 45870)
    },
    'Gradient Boosting': {
        'SMOTE': (3012, 4160, 9036, 66266),
        'Under Sampling': (8522, 21915, 3652, 48385)
    },
    'Logistic Regression': {
        'SMOTE': (7999, 21627, 3940, 48908),
        'Under Sampling': (7886, 21321, 4246, 49021)
    },
    'Linear SVM': {
        'SMOTE': (7943, 21475, 4092, 48964),
        'Under Sampling': (7886, 21321, 4246, 49021)
    }
}

models = list(conf_matrices.keys())


fig, axes = plt.subplots(5, 2, figsize=(16, 22))
for i, model in enumerate(models):
    for j, method in enumerate(['SMOTE', 'Under Sampling']):
        tp, fp, fn, tn = conf_matrices[model][method]
        cm = np.array([[tp, fp], [fn, tn]])
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Diabetic", "Non-Diabetic"])
        ax = axes[i, j]
        disp.plot(ax=ax, values_format='d')
        ax.set_title(f'{model} - {method}')

plt.tight_layout()
plt.suptitle("Confusion Matrices for All Models (SMOTE vs. Under Sampling)", fontsize=18, y=1.02)
plt.show()
```

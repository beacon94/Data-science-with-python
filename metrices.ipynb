```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for model performance (SMOTE vs. Under Sampling)
models = ['Decision Tree', 'Random Forest', 'Gradient Boosting', 'Logistic Regression', 'Linear SVM']
smote_data = {
    'Accuracy':  [0.77, 0.84, 0.84, 0.69, 0.69],
    'Precision': [0.25, 0.37, 0.42, 0.27, 0.27],
    'Recall':    [0.28, 0.14, 0.25, 0.67, 0.66],
    'F1-Score':  [0.26, 0.21, 0.31, 0.38, 0.39],
    'AUC-ROC':   [0.57, 0.55, 0.60, 0.68, 0.68]
}
under_data = {
    'Accuracy':  [0.61, 0.66, 0.69, 0.69, 0.69],
    'Precision': [0.21, 0.26, 0.28, 0.27, 0.27],
    'Recall':    [0.58, 0.70, 0.70, 0.65, 0.65],
    'F1-Score':  [0.30, 0.38, 0.40, 0.38, 0.38],
    'AUC-ROC':  [0.60, 0.68, 0.70, 0.67, 0.67]
}

# Convert to DataFrames
df_smote = pd.DataFrame(smote_data, index=models)
df_under = pd.DataFrame(under_data, index=models)

# Plot Performance Metrics for SMOTE and Under Sampling
fig, axes = plt.subplots(3, 2, figsize=(14, 15))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC']
for i, metric in enumerate(metrics):
    ax = axes[i // 2, i % 2]
    ax.plot(models, df_smote[metric], label='SMOTE', marker='o')
    ax.plot(models, df_under[metric], label='Under Sampling', marker='o')
    ax.set_title(f'{metric} Comparison')
    ax.set_ylabel(metric)
    ax.set_xticklabels(models, rotation=45)
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.suptitle('Model Performance Comparison (SMOTE vs. Under Sampling)', fontsize=16, y=1.02)
plt.show()
```
